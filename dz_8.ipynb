```python
#1
import numpy as np
with open("input.txt",encoding = "utf8") as file:
    arr = [list(map(float, line.split())) for line in file]
    arr = np.array(arr)
arr = arr.T
print(arr)
```

```python
#2
import numpy as np

arr = np.loadtxt("input.txt", dtype=float)


median_val = np.median(arr)
mean_val = np.mean(arr)
std_val = np.std(arr)

print("{:.2f}".format(median_val),"{:.2f}".format(mean_val),"{:.2f}".format(std_val))
```

```python
#3
import numpy as np

arr = np.loadtxt('input.csv', delimiter=',')

arr[1::2, ::2] /= 2 
arr[::2, 1::2] /= 2    

np.savetxt('output.csv', arr, fmt='%g', delimiter=',')
```

```python
#4
import numpy as np

arr = np.loadtxt('input.csv', delimiter=',')
A  = np.sum(arr, axis=0)
print(np.argmax(A))
```

```python
#5
import numpy as np
arr = np.loadtxt('input.csv', delimiter=',')
A = np.std(arr,axis = 1)
count1 = 0
count2 = 0
for i in range(len(A)):
    if abs(A[i])>4:
        count2+=1
    else:
        count1+=1
if count1>count2:
    print(1)
else:
    print(2)
```
