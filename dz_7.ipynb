```python
#4
import numpy as np
with open("input.txt",encoding = "utf8") as file:
    a = [line.strip() for line in file]
    start = float(a[0])
    end = float(a[1])
    count = int(a[2])
a = np.linspace(start,end,count)
a[::7] /=3
a[4::7] *=2
file = open("output.txt", encoding = "utf8",mode = 'w')
for x in a:
    file.write(f'{x:.2f}\n')
file.close()
```

```python
#5
import math
def f(x):
    return math.sin(math.tan(1+x/1000))
    
def solver(left_boundary, right_boundary):
    supposed_error = 1.0e-8
    previous_middle = left_boundary
    k = 0;
    while right_boundary - left_boundary > 2 * supposed_error or k <100:
        middle = (right_boundary + left_boundary) / 2
        if abs(f(middle)) < 1.0e-16:
            return middle
        if abs(f(previous_middle) - f(middle)) < 1.0e-7:
            return middle
            
        if f(left_boundary) * f(middle) < 0:
            right_boundary = middle
        else:
            left_boundary = middle
        previous_middle = middle
        k+=1
    return (right_boundary + left_boundary) / 2
           
    return (right_boundary + left_boundary) / 2
n = int(input())
for _ in range(n):
    a, b = map(int,input().split())
    root = solver(a,b)
    print(f"{root:.6F}")
```
