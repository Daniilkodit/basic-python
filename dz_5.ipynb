```python
#4
f_in = open("input.txt", mode = 'r')
f_out = open("output.txt", mode = 'w')
arr = f_in.readlines();
arr = [line.split() for line in arr]
n = len(arr)
m = len(arr[0])
arr_t = [[arr[j][i] for j in range(n)] for i in range(m)]
for line in arr_t:
    f_out.write(" ".join(line).strip() +"\n");
```

```python
#5
f_in = open("input.txt", mode = 'r')
f_out = open("output.txt", mode = 'w')
arr = f_in.readlines();
arr =[line.split('\t') for line in arr]
arr.sort(key = lambda line: int(line[1]),reverse = True)
A = [el for el in arr if el[2] =='A']
B = [el for el in arr if el[2] =='B']
for line in A:
    f_out.write("\t".join(line).strip() +"\n");
for line in B:
    f_out.write("\t".join(line).strip() +"\n");
```
