```python
#1
class Incapsulator:
    def __init__(self,value = 0):
        self.value = value
    def get_value(self):
        return self.value
    def set_value(self,value):
        self.value = value
a = Incapsulator()
a.set_value(-1.0)
print(a.get_value())
```

```python
#2
class Student:
    def __init__(self,name,age):
        self._age = float(age)
        self._name = name
    def __str__(self):
        return "Student " + self._name + " of age " + str(self._age)
    def __repr__(self):
        return f"Student({self._name}, {self._age})"
```
