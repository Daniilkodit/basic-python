```python
#4
s = input();
a = ['-', ')', '(',' ']
for i in a:
    s = s.replace(i,'')
print(s)
```

```python
#5
s = input()
s_list = list(s)  

for i in range(len(s_list)):
    if 65 <= ord(s_list[i]) <= 90: 
        new_char = chr((ord(s_list[i]) + 3 - 65) % 26 + 65)
        s_list[i] = new_char
    elif 97 <= ord(s_list[i]) <= 122:
        new_char = chr((ord(s_list[i]) + 3 - 97) % 26 + 97)
        s_list[i] = new_char

s = ''.join(s_list)
print(s)
```
