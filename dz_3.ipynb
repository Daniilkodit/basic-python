```python
#1
file = open('input.txt',mode ='r')
s = file.read().split()
s = map(lambda word: word[::-1],s)
print(' '.join(s))
```

```python
#2
file = open('input.txt',mode = 'r',encoding='utf-8')
s = file.read()
prep = ['.',',','!','?']
for char in prep:
    s = s.replace(char,' ')
s = s.split()
for word in s:
    if(word.find('ё')!=-1):
        print(word)
```

```python
#3
file = open('input.txt',mode = 'r',encoding='utf-8')
lines = file.read().split('\n')
for line in lines:
    arr = line.split()
    print(*arr[::2])
```

```python
#4
file_in = open('input.txt',mode = 'r',encoding='utf-8')
file_out = open('output.txt', mode = 'w',encoding = 'utf-8')
s = file_in.read().split('\n')
N = int(s[0])
names = [s[i] for i in range(1,N+1)]
dates = [s[i] for i in range(N+1,2*N+1)]
for name,date in zip(names,dates):
    file_out.write(name +'\t'+date+'\n')
file_in.close()
file_out.close()
```

```python
#5
import itertools as it
file = open('input.txt',mode = 'r',encoding='utf-8')
s = file.read().split()
for t in it.combinations(s, 3):
    print('1: '+t[0] +' 2: '+t[1] +' 3: '+t[2])
```
