```python
def System(n,base):
    if(n<base):
        print(n,end='')
        return
    System(n//base,base)
    print(n%base,end='')
        
        
n = int(input())
base = 2
System(n,base)
```

```python
import math
limit,N = map(int,input().split())
fig_sum = 0;
for i in range(N):
    line = input().split()
    line_min = math.inf;
    for element in line:
        price,count = map(int,element.split('/'))
        if price <= limit and price*count<line_min:
            line_min = price*count
    if(math.isinf(line_min)==False):
        fig_sum+= line_min
print(fig_sum)
```
