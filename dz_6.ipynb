```python
#1
from collections import Counter
s = input();
the_dict = Counter(s[i] for i in range(len(s)))
answer = min(the_dict[key] for key in the_dict)
answer = min(key for key in the_dict if the_dict[key] == answer)
print(answer)
```

```python
#2
with open("input.txt",encoding = "utf8") as file:
    animal_set = set()
    for line in file:
        animal = line.split()
        animal_set.add(animal[1])
animal_list = list(animal_set)
animal_list.sort(key = len)
print(*animal_list,sep = '\n')
```

```python
#3
with open("input.txt",encoding = "utf8") as file:
    animal_dict = {"female" : set(),"male" :set() }
    for line in file:
        animal = line.split()
        animal_dict[(animal[2].strip())].add(animal[1])
animal_list = [animal for animal in animal_dict["male"] if animal in animal_dict["female"]]

animal_list.sort(key = len)
if animal_list:
    print(*animal_list,sep='\n')
else:
    print(0)
```

```python
#4
from collections import Counter
with open("input.txt",encoding = "utf8") as file:
     animal_dict = Counter(line.split()[1] for line in file)
animal_list = [[key,animal_dict[key]] for key in animal_dict]
animal_list.sort(key = lambda animal: animal[1],reverse = True)
for animal in animal_list:
    print(animal[0]+" - "+str(animal[1]))
```

```python
#5
with open("input.txt",encoding = "utf8") as file:
    animal_dict = dict()
    for line in file:
        animal = line.split()
        if animal[1]  not in animal_dict:
            animal_dict[animal[1]] = list()
        animal_dict[animal[1]].append(animal[0])
animal_list = [[key,animal_dict[key]] for key in animal_dict]
animal_list.sort(key = lambda animal: len(animal[0]))
for animal in animal_list:
    ID = animal[1]
    ID.sort(key = lambda a: int(a))
    print(animal[0] + ': ',end='')
    print(*ID,sep = ', ')
```
